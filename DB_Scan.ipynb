## Density Based Hierarchy

DB Scan Alogorithm is used in following cases:

- Used in the case of bunch of outliers.
- In the case of irregular shape.(Understanding shape using Scatter plot)
- In case of kmeans the shape is spherical.

Contains two parameters :

1. EPS : Epsilon 
2. min point : No of points required in the circle.

Point types :

1. Core point : If Min points present in the Epsilon(Circle).
2. Boundary point : Min points are not present in Epsilon.
3. Noise Point : Only single point is present within the circle(Epsilon).


### Algorithm : Anomaly detection

- It detects abnormal requests.
- Ex : Email Notifications (Someone is trying to login to your gmail).

Isolation Forest :
- Detect Abnormal point.
- Isolation technique helps us to find out whether the data point is normal or abnormal.
- Grow the decision tree until we reach the  all the data points as leaf nodes

Isoalted point or Abnormal Point:
With minimum number of splits we reach the leaf node

Path length: 
- No.of splits required to convert the data point into leaf node
- Here we are creating multiple trees unlike decison tree we dont use information gain to define the root node at each level

##### Why required isolation tree :

1. Formula :

- Anomaly score = 1 - Anomaly(Strong).
- Anomaly Score = 0.5 - Anomaly.
- Anomaly Score = 0 - Normal Data point.
- S(X,n)=2**-E|h(X)|/c(n)
    - h(X) : Path length of Data point.
    - E|h(X)| : Average path length of data point.
    - C(n): Average path of tree.

    Average path of tree (C(n)): (2*h(n-1))**(-2*(n-1)/n)
    - n : no of data points.
    - h : harmonic numbers 

    E|h(X)| : 
    - If E|h(X)| << (C(n)):
        Anamoly -> Score is close to 1.

## Algorithm Steps :
- For Each tree :
1. Randomly select n samples
2. Build tree using random split
3. Compute path length for each data point

After building all trees :
1. Compute avg path length
2. Noralise using C(n) - Average path of tree 
3. Compute anomaly score
4. Classify using cotimination threshold

For harmonic calculations : 
- If n is small : Calculate directly = 1+1/2+1/3+...+1/n
- If n is large : Use Approximation H(n)~log(n)+0.577
